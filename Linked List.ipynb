# Linked List

```python
'''
Implementation of Linked List in Python
'''

class Node:
    def __init__(self, x, next_node=None, prev_node=None):
        self.val = x
        self.next_node = next_node # default next node is none, we use set_next to create a new link
        self.prev_node = prev_node
        
    def get_data(self):
        return self.val # return the value stored at node
    
    def get_next(self):
        return self.next_node # find the next node
    
    def set_next(self, new_next):
        self.next_node = new_next

class MyLinkedList:

    def __init__(self, size=0, head=None, tail=None):
        """
        Initialize your data structure here.
        """
        self.size = size
        self.head = head # when first initalized, head is none
        self.tail = tail # tail is none when first initialized

    def get(self, index):
        """
        Get the value of the index-th node in the linked list. If the index is invalid, return -1.
        """
        if index > self.size or index < 0:
            return -1
        cur = self.head
        ind_ = index
        while ind_ > 0:
            cur = cur.next_node # loop through until we arrive at the right index
            ind_ -= 1
        return cur.val
        
    # recall, to add at head, we first link the new node to head, then assign the new node as the head
    def addAtHead(self, val):
        """
        Add a node of value val before the first element of the linked list. After the insertion, the new node will be the first node of the linked list.
        """
        if self.head = None: # if there is not a head
            new_head = Node(val) # creates a new node using value we want to insert
            self.head = new_head # head node initialized as inserted node for our linked list
        else:
            new_head = Node(val) # create new_head as a node object with val
            new_head.next_node = self.head # links to prev head as next_node
            self.head = new_head # reassigns head
        self.size += 1 # increase size to match
        
        

    def addAtTail(self, val):
        """
        Append a node of value val to the last element of the linked list.
        """
        if self.head = None:
            new_head = Node(val) # covers case at beginning where we first initialize linked list
            self.head = new_head
        else:
            cur = self.head
            while cur.next_node:
                cur = cur.next_node
            new_tail = Node(val)
            cur.next_node = new_tail # after while loop, cur reaches tail node -> here we link new_tail as the new tail to linked list
        self.size += 1
        

    def addAtIndex(self, index, val):
        """
        Add a node of value val before the index-th node in the linked list. If index equals to the length of linked list, the node will be appended to the end of linked list. If index is greater than the length, the node will not be inserted.
        """
        if self.size < index < -2:
            pass
        elif index == self.size:
            self.addAtTail(val)
        elif index == 0 or index == -1:
            self.addAtHead(val)
            
        cur = self.head
        ind_ = index - 1
        while ind_ > 0:
            cur = cur.next_node
            ind_ -= 1
        new_node = Node(val)
        new_node.next_node = cur.next_node # reassigns to new_node
        cur.next_node = new_node
        self.size += 1
        
        

    def deleteAtIndex(self, index):
        """
        Delete the index-th node in the linked list, if the index is valid.
        """
        if self.size - 1 < index < 0:
            pass
        if index == 0:
            self.head = (self.head).next_node # assign next node as new head
            self.size -= 1
        
        cur = self.head
        ind_ = index - 1
        while ind_ > 0:
            cur = cur.next_node
            ind_ -= 1
        if cur.next_node:
            cur.next_node = (cur.next_node).next_node # reassigns node at index's next_node as next-next node
        else:
            cur.next_node = None # if we've deleted the tail
        self.size -= 1
        

# Your MyLinkedList object will be instantiated and called as such:
# obj = MyLinkedList()
# param_1 = obj.get(index)
# obj.addAtHead(val)
# obj.addAtTail(val)
# obj.addAtIndex(index,val)
# obj.deleteAtIndex(index)
```
